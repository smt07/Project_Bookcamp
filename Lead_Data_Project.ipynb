```python
def calculate_mi(series):
    return mutual_info_score(series, df_train_full.converted)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI').round(5)


display(df_mi.head())
display(df_mi.tail())
```

**One-hot encoding**


```python
from sklearn.feature_extraction import DictVectorizer
train_dict = df_train[categorical_most + numerical].to_dict(orient='records')
train_dict[0]
```

```python
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
X_train.shape
```

```python
dv.feature_names_
```

**Training logistic regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
model.predict_proba(X_val)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
```

```python
y_pred
```

```python
converted = y_pred > 0.5
```
